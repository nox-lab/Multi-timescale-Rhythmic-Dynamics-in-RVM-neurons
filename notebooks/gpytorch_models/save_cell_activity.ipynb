## Saving cell activity for faster access

This notebook imports all of the data and stores the spikes for faster access without having to import the entire dataset. The files and the spikes need to be saved in the same order so that later functions which rely on importing the cells in the right order behave properly.

```python
import pickle
from quantities import s
from rvm_analysis.utils import import_spontaneous_cells, import_neutral_cells, import_neutral_cells_extra
from rvm_analysis.data_loaders import read_dataset_by_protocol
```

```python
def deprecated_loading_using_spike2():
    dm = import_spontaneous_cells()
    dm2 = import_neutral_cells()
    dm3 = import_neutral_cells_extra()
    return dm,dm2,dm3

dm,dm2,dm3 = deprecated_loading_using_spike2()

dm_nwb = read_dataset_by_protocol(['ongoing'])

dm_extra_neut_nwb = read_dataset_by_protocol(['evoked/ongoing'])
```

```python
print(len(list(dm.spiketrain_iterator(cell_type='All'))))
print(len(list(dm_nwb.spiketrain_iterator(cell_type='ON'))))
print(len(list(dm_nwb.spiketrain_iterator(cell_type='OFF'))))

print("-----")

print(len(list(dm3.spiketrain_iterator(cell_type='All'))))
print(len(list(dm_nwb.spiketrain_iterator(cell_type='Neutral'))))

print("----")
print(len(list(dm2.spiketrain_iterator(cell_type='All'))))
print(len(list(dm_extra_neut_nwb.spiketrain_iterator(cell_type='Neutral'))))
```

```python
def deprecated_save_code_spike2():
    spikes_ON = list(dm.spiketrain_iterator(cell_type='ON'))
    spikes_OFF = list(dm.spiketrain_iterator(cell_type='OFF'))

    cut_spikes_ON = dm.get_event_windows(size=25*60*s,cell_type='ON',zero=False,event_centre=False)
    cut_spikes_OFF = dm.get_event_windows(size=25*60*s,cell_type='OFF',zero=False,event_centre=False)
    cut_spikes_NEUTRAL_extra = dm3.get_event_windows(size=1500*s,cell_type='All',zero=False,event_centre=False)
    spikes_NEUTRAL = list(dm2.spiketrain_iterator(cell_type='All'))
    spikes_NEUTRAL_extra = list(dm3.spiketrain_iterator(cell_type='All'))

    print(spikes_NEUTRAL)
    print(cut_spikes_NEUTRAL_extra)

    files_ON = dm.get_files(cell_type="ON")
    files_OFF = dm.get_files(cell_type="OFF")
    files_NEUTRAL = dm2.get_files(cell_type="All")
    files_NEUTRAL_extra = dm3.get_files(cell_type="All")
    print(files_ON,files_OFF)
```

```python
spikes_ON = list(dm_nwb.spiketrain_iterator(cell_type='ON'))
spikes_OFF = list(dm_nwb.spiketrain_iterator(cell_type='OFF'))

cut_spikes_ON = dm_nwb.get_event_windows(size=25*60*s,cell_type='ON',zero=False,event_centre=False)
cut_spikes_OFF = dm_nwb.get_event_windows(size=25*60*s,cell_type='OFF',zero=False,event_centre=False)
cut_spikes_NEUTRAL_extra = dm_nwb.get_event_windows(size=1500*s,cell_type='Neutral',zero=False,event_centre=False)
spikes_NEUTRAL = list(dm_extra_neut_nwb.spiketrain_iterator(cell_type='Neutral'))
spikes_NEUTRAL_extra = list(dm_nwb.spiketrain_iterator(cell_type='Neutral'))

print(spikes_NEUTRAL)
print(cut_spikes_NEUTRAL_extra)

files_ON = dm_nwb.get_files(cell_type="ON")
files_OFF = dm_nwb.get_files(cell_type="OFF")
files_NEUTRAL = dm_extra_neut_nwb.get_files(cell_type="Neutral")
files_NEUTRAL_extra = dm_nwb.get_files(cell_type="Neutral")
print(files_ON,files_OFF)
```

```python
def save_cut_spikes():
    with open("../../CUT_CELL_ACTIVITIES_unzeroed.pkl", "wb") as file:
        pickle.dump((cut_spikes_ON, cut_spikes_OFF, spikes_NEUTRAL, cut_spikes_NEUTRAL_extra), file)
def save_spikes():
    with open("../../UNCUT_CELL_ACTIVITIES.pkl", "wb") as file:
        pickle.dump((spikes_ON, spikes_OFF, spikes_NEUTRAL, spikes_NEUTRAL_extra), file)
def save_files():
    with open("../../saved_files_new.pkl", "wb") as file:
        pickle.dump((files_ON,files_OFF,files_NEUTRAL,files_NEUTRAL_extra), file)
```

```python
save_spikes()
save_cut_spikes()
save_files()
```
